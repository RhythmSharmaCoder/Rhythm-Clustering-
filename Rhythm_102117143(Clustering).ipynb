**K-MEANS CLUSTERING**

```python
kMeanClusteringParameters = setup(data, verbose=False) #Parameter-Setting
```

```python
KMeanClusteringModel = create_model('kmeans', num_clusters=3)
```

```python
KMeanClusteringModel = create_model('kmeans', num_clusters=4)
```

```python
KMeanClusteringModel = create_model('kmeans', num_clusters=5)
```

```python
num_clusters = [3, 4, 5]
silh = []
calinski = []
davies = []
```

```python
def get_parameters(num_clusters):
    KMeanClusteringModel = create_model('kmeans', num_clusters=num_clusters)
    prediction = assign_model(KMeanClusteringModel)
    performance = pull()
    silh.append(performance['Silhouette'][0])
    calinski.append(performance['Calinski-Harabasz'][0])
    davies.append(performance['Davies-Bouldin'][0])

def results():
  for num in num_clusters:
    get_parameters(num)

print("For No Data Preprocessing")
results()
results_no_preprocessing = [silh, calinski, davies]
print("For Normalisation (N)")
normparam = setup(data=data,normalize = True, normalize_method = 'zscore', verbose=False)
results()
print("For Transformation (T)")
transformparam=setup(data=data,transformation = True, transformation_method = 'yeo-johnson', verbose=False)
results()
print("For PCA")
pcaparam=setup(data=data,transformation = True,pca = True, pca_method = 'linear', verbose=False)
results()
print("For N+T")
ntparam=setup(data=data,transformation = True, normalize = True,normalize_method = 'zscore',
      transformation_method = 'yeo-johnson', verbose=False)
results()
print("For N+T+PCA")
ntpcaparam=setup(data=data,transformation = True, normalize = True, pca = True,
      normalize_method = 'zscore', transformation_method = 'yeo-johnson',
      pca_method = 'linear',  verbose=False)
results()
```

```python
pt.field_names=["Parameters","No data preprocessing c=3","No data preprocessing c=4 ","No data preprocessing c=5","Using Normalisation c=3","Using Normalisation c=4","Using Normalisation c=5","Using Transformation c=3","Using Transformation c=4","Using Transformation c=5","Using PCA c=3","Using PCA c=4","Using PCA c=5","Using N+T c=3","Using N+T c=4","Using N+T c=5","Using N+T+PCA c=3","Using N+T+PCA c=4","Using N+T+PCA c=5"]
pt.add_row(["Silhouette",*silh])
pt.add_row(["Calinski-Harabasz",*calinski])
pt.add_row(["Davies-Bouldin",*davies])
print("USING K-MEANS CLUSTERING")
print(pt)
```

Looks like number of clusters=3 works best and accuracy is increasing mostly with using Transformation and PCA.

```python
import csv

with open("kmeans_output.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(pt.field_names)
    for row in pt._rows:
        writer.writerow(row)
```

```python
visualise_kmeans=setup(data,transformation = True, transformation_method = 'yeo-johnson', verbose=False)
KMeanModel = create_model('kmeans', num_clusters=3)
evaluate_model(KMeanModel)
plot_model(KMeanModel, plot='cluster')
plot_model(KMeanModel, plot='tsne')
plot_model(KMeanModel, plot = 'distribution')
```
